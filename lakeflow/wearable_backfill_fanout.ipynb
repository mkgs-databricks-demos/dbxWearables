with more conservative rate-limit pacing.

```python
DEFAULT_SECRET_SCOPE = "dbxw_zerobus_credentials"

try:
    dbutils.widgets.text("secret_scope", DEFAULT_SECRET_SCOPE, "Secret Scope")
    dbutils.widgets.text("providers", "garmin", "Providers (comma-separated, blank=all)")
    dbutils.widgets.text("user_ids", "", "User IDs (comma-separated, blank=all enrolled)")
    dbutils.widgets.text("start_date", "", "Start date (YYYY-MM-DD)")
    dbutils.widgets.text("end_date", "", "End date (YYYY-MM-DD)")
    dbutils.widgets.text("delay_seconds", "3", "Delay between dates (seconds)")

    secret_scope = dbutils.widgets.get("secret_scope")
    providers_arg = dbutils.widgets.get("providers").strip()
    user_ids_arg = dbutils.widgets.get("user_ids").strip()
    start_date_str = dbutils.widgets.get("start_date").strip()
    end_date_str = dbutils.widgets.get("end_date").strip()
    delay_seconds = float(dbutils.widgets.get("delay_seconds"))
except Exception:
    secret_scope = DEFAULT_SECRET_SCOPE
    providers_arg = "garmin"
    user_ids_arg = ""
    start_date_str = ""
    end_date_str = ""
    delay_seconds = 3.0

assert start_date_str and end_date_str, "start_date and end_date are required for backfill"
```
